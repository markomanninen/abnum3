## Process Greek Corpora

First task is to get a big raw text to operate with. CLTK provides an importer to the Perseus open source data source. Other good option would be to use TLG: http://opengreekandlatin.github.io/First1KGreek/

```python
import sys
sys.version
```

```python
#pip install cltk
from cltk.corpus.utils.importer import CorpusImporter
corpus_importer = CorpusImporter('greek')
corpus_importer.list_corpora
```

```python
corpus = "greek_text_perseus"
#corpus = "greek_software_tlgu"
corpus_importer.import_corpus(corpus)
```

```python
# download greek betacode decoder and import:
# https://raw.githubusercontent.com/epilanthanomai/hexameter/master/betacode.py
import os, re, betacode
from xml.dom import minidom
from os.path import expanduser
from collections import Counter
#from abnum import Abnum, greek
from abnum3.abnum import Abnum, greek
g = Abnum(greek)

home = expanduser("~")

# directory: by default corpora is downloaded to the user root under cltk_data
dirt = home + '\\cltk_data\\greek\\text\\' + corpus
dire = dirt + '\\%s\\opensource\\'

all_greek_text_file = "all_greek_texts.txt"
greek_corpora = {}

def betadecode(x):
    return betacode.betacode_to_unicode(x)

def remove_tags(x):
    return betadecode(re.sub('<[^<]+?>', '', x))

def add_line_to_corpora(item, corp, i):
    line = item.toxml().strip()
    if line:
        line = ' '.join(filter(lambda x: x.strip() != "", remove_tags(line).split()))
        if line:
            greek_corpora[corp][i]['content'].append(line)

def append_to_file(file, string):
    with open(file, 'a', encoding="utf-8") as f:
        f.write(string)

# tags used to retrieve lines from xml file
tags = {
    "Aeschines": ["div1", ["head", "p"]],
    "Aeschylus": ["div1", ["speaker", "l"]],
    "Andocides": ["div1", ["head", "p"]],
    "Anth": ["div1", ["label", "p"]],
    "Apollodorus": ["div1", ["p"]],
    "Apollonius": ["div1", ["l"]],
    "Appian": ["p", []],
    "Aratus": ["p", []],
    "Aretaeus": ["p", ["foreign"]],
    "Aristides": ["body", ["head", "p"]],
    "Aristophanes": ["div1", ["speaker", "l"]],
    "Aristotle": ["p", []],
    "Arrian": ["body", ["p"]],
    "Athenaeus": ["body", ["p"]],
    "Bacchylides": ["div2", ["persName", "l"]],
    "Bible": ["div1", ["head", "p"]],
    "Callimachus": ["div1", ["head", "p"]],
    "Colluthus": ["p", []],
    "Demades": ["p", []],
    "Demosthenes": ["p", []],
    "Diodorus": ["p", []],
    "Diogenes": ["div1", ["head", "p"]],
    "Dionysius": ["p", []],
    "Dinarchus": ["p", []],
    "DioChrys": ["p", []],
    "Epictetus": ["p", []],
    "Euclid": ["p", []],
    "Galen": ["p", []],
    "Herodotus": ["p", []],
    "Hippocrates": ["p", []],
    "Hyperides": ["p", []],
    "JebbOrators": ["foreign", []],
    "Josephus": ["p", []],
    "Lucian": ["p", []],
    "Lycophron": ["p", []],
    "Lycurgus": ["div1", ["head", "p"]],
    "Lysias": ["div1", ["head", "p"]],
    "Oppian": ["p", []],
    "Nonnos": ["p", []],
    "Plato": ["p", []],
    "Plutarch": ["p", []],
    "Polybius": ["p", []],
    "Sibyl": ["div", ["head", "p"]],
    "Tryphiodorus": ["p", []],
    "Xenophon": ["p", []],
    "Lycophron": ["p", []],
    "Theophrastus": ["p", []],
}

# filter dirs and files to get all greek authors
authors = list(filter(lambda x: "." not in x and "json" not in x, os.listdir(dirt)))

# filter only greek xml files
def is_greek_file(x):
    return ("greek_orators" in x or \
            "attic_orators" in x or \
            "_01.xml" in x or "_02.xml" in x or \
            "_03.xml" in x or "_gk.xml" in x) and ".json" not in x

# init corpora data var
for author in authors:
    works = list(filter(is_greek_file, os.listdir(dire % author)))
    if works:
        greek_corpora[author] = []
        for work in works:
            greek_corpora[author].append({
                "tag": tags[author] if author in tags else ["l", []],
                "file": work,
                "content": [],
                "simplified": "",
                "uwords": {},
                "length": 0})
```

```python
try:
    os.remove(all_greek_text_file)
except OSError:
    pass

# retrieve corpora texts
for corp, items in greek_corpora.items():
    if items:
        for i, data in enumerate(items):
            xmldoc = minidom.parse((dire % corp) + data['file'])
            if data['tag'][0] == "foreign":
                for item in xmldoc.getElementsByTagName("foreign"):
                    if item.hasAttribute("lang") and item.getAttribute("lang") == "greek":
                        add_line_to_corpora(item, corp, i)
            else:
                itemlist = xmldoc.getElementsByTagName(data['tag'][0])
                for item in itemlist:
                    if len(data['tag'][1]):
                        for tag in data['tag'][1]:
                            for item2 in item.getElementsByTagName(tag):
                                add_line_to_corpora(item2, corp, i)
                    else:
                        add_line_to_corpora(item, corp, i)
            greek_corpora[corp][i]['simplified'] = g.preprocess(' '.join(greek_corpora[corp][i]['content'])).upper()
            greek_corpora[corp][i]['uwords'] = Counter(greek_corpora[corp][i]['simplified'].split())
            greek_corpora[corp][i]['length'] = len(greek_corpora[corp][i]['simplified'].replace(" ", ""))
            greek_corpora[corp][i]['lperw'] = greek_corpora[corp][i]['length'] / len(greek_corpora[corp][i]['uwords'])
            append_to_file(all_greek_text_file, greek_corpora[corp][i]['simplified'] + "\n")
            print(corp, data['file'], greek_corpora[corp][i]['length'])
    else:
        print("no corpora: " + corp)
```

## Solve the riddle

Using parameters from the riddle, lets try to solve it by brute force.

- knowing isopsephical value 1697
- knowing the length of the word
- knowing the details of the syllables of the word
- comparing the context of the result

### Apply isopsephical value to the unique words

Start by adding isopsephical value to the words as a new column:

```python
df[4] = df[0].apply(lambda x: g.value(x))
df
```

How many words there are with the isopsephical value 1697 and letters counting nine?

```python
# retrieve the words having nine letters and their value as 1697
df[df[4] == 1697][df[0].str.len() == 9]
```

It turns out that there are just very few words meeting the criteria of the riddle. We could already make analyzing of the words manually.

The last interest is to split words to syllables so that we can make the final conclusion. I will however add one extension to the search and allow the count of the letters to be between 8 and 10. That is due to double consonant and long / short vowelspecialty of the Greek language.

```python
# we might want to make another filter by using syllabification for the word and count
# only those words having the length between 8 and 10 (short and long vowels included)
from greek_accentuation.syllabify import syllabify

c = df[df[4] == 1697][df[0].str.len() < 11][df[0].str.len() > 7]
# add syllabified column
c[5] = c[0].apply(lambda x: syllabify(x))
# add length of the syllables column
c[6] = c[5].apply(lambda x: len(x))
c[c[6] == 4]
```

```python
# more strictly only words with 9 letters and 4 syllables
c[c[6] == 4][c[3] == 9]
```

```python
# words having 9 letters, 4 syllables where the first three of the syllables has two letters, last one the rest 3
c[c[6] == 4][c[3] == 9][c.apply(lambda x: len(x[5][0]) == 2 and len(x[5][1]) == 2 and len(x[5][2]) == 2, axis=1)]
```

Finally there is the rule of five consonants in the word in the original riddle. Lets refactor all of this and the previous ones to the single callable function and apply it to dataframe.

```python
# the word should have 5 mutes ie consonants
consonants = "ΨΖΞΒΦΧΘΓΔΜΛΚΠΡΣΤΝ"
def has_five_mutes(x):
    tot = 0
    for c in consonants:
        tot += x[0].count(c)
    return tot == 5
# the word should have four syllables
def has_four_syllables(x):
    return x[6] == 4
# the word should have two letters in the first three syllables, and the rest (3) letters in the last
def has_two_letters_in_first_three_syllables(x):
    return len(x[5][0]) == 2 and len(x[5][1]) == 2 and len(x[5][2]) == 2
# this makes nine letters in total
def has_nine_letters(x):
    return len(x[0]) == 9
# riddle wrapper function
def riddle(x):
    return has_nine_letters(x) and has_five_mutes(x) and has_four_syllables(x) and has_two_letters_in_first_three_syllables(x)
```

```python
# solve the riddle
c[c.apply(riddle, axis=1)]
```

Thus we have found one exact match for the riddle. The word is:

# ΑΜΦΕΚΑΛΥΨ

(amphekalyps / amfecalyps) meaning all covering.
